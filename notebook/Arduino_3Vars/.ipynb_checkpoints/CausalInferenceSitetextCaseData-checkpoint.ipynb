```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
def testpic(S1,V0,V1,tit):
    # Define some plt properties
    fig = plt.figure(figsize = (9,9))
    fig.subplots_adjust(hspace=0.5, wspace=0.5)
    plt.rcParams['font.size']=8
    ax = fig.add_subplot(1,1,1, projection='3d')
    ax.scatter3D(V0, V1, S1, zdir='z', color='green', marker='.', s=100)
    ax.scatter3D(V0[41:49], V1[41:49], S1[41:49], zdir='z', color='red',marker='.', s=100)
    ax.set_xlabel('V0'), ax.set_ylabel('V1'),ax.set_zlabel('S1')
    ax.set_xlim(0,5), ax.set_ylim(0,5),ax.set_zlim(0,1)
    ax.view_init(elev=60, azim=50)#, roll=0)
    plt.title(tit)
    
    print('correlation matrix:')
    print(pd.DataFrame(np.corrcoef([S1, V0, V1]), index=['S1','V0','V1'], columns=['S1','V0','V1']) )
```

```python
def scm(cause,effect):
    model = LinearRegression().fit(cause, effect)
    model.score(cause, effect)
    model.intercept_
    model.coef_

    print ('linear structural equation: ')
    print ('effect = ',np.round(model.coef_,2),'* [cause].T + ',round(model.intercept_,2), ' coefficient of determination: ',round(model.score(cause, effect),2) )
```
